```python
import os
from os import path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dirname = os.path.abspath(os.curdir) + r'/data/underwork/5/'

tsdf_c = pd.read_csv(path.join(os.sep, dirname, 'calm_p.csv'))
tsdf_c.set_index('Time').sort_index()

passangers = pd.read_csv('data/passengers.csv')
passangers = pd.read_csv('data/passengers.csv')
passangers.head()
```

```python
passangers.describe()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(15, 10))

axs[0].plot(tsdf_c['1'])
axs[0].set_xlabel('Time')
axs[0].set_ylabel('mV')
axs[0].set_title('Покой. Время')

axs[1].plot(passangers['Passengers'])
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Qty')
axs[1].set_title('Пассажиры')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
rcParams['figure.figsize'] = 11, 9

decompose = seasonal_decompose(passangers['Passengers'], period=10)
decompose.plot()
plt.show()
```
