```python
import numpy as np
import matplotlib.pyplot as plt
```

Параметры

```python
num_samples = 20000
duration = 10
mean = 1
sd = 2
```

Белый шум (база)

```python
time_axis = np.linspace(0, duration, num_samples)
white_noise_y = np.random.normal(mean, sd, num_samples)
```

```python
plt.figure(figsize=(20, 6))

plt.plot(time_axis, white_noise_y)

plt.title('Белый шум')
plt.xlabel('Время (секунды)')
plt.ylabel('Амплитуда')
```

Розовый шум

```python
def gen_pink_noice(base_noice):
    rfft = np.fft.rfft(base_noice)
    freq = np.fft.rfftfreq(num_samples, 1.0 / num_samples) + 1
    return  np.fft.irfft(rfft / np.sqrt(freq))
```

```python
pink_noise = gen_pink_noice(white_noise_y)
```

```python
plt.figure(figsize=(20, 6))  

plt.plot(time_axis, pink_noise)
plt.title('Розовый шум (свертка белого шума)')
plt.xlabel('Время (секунды)')
plt.ylabel('Амплитуда')

plt.show()
```
